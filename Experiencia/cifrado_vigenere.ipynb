```python
import pandas as pd



alfabeto = "abcdefghijklmnñopqrstuvwxyz"

# funciones vigenere

def expandir_clave(clave, longitud):
    return (clave.lower() * (longitud // len(clave) + 1))[:longitud]

def cifrar_vigenere(texto, clave):
    texto = texto.lower()
    clave_ext = expandir_clave(clave, len(texto))
    resultado = ""

    for t, k in zip(texto, clave_ext):
        if t in alfabeto:
            t_index = alfabeto.index(t)
            k_index = alfabeto.index(k)
            nuevo_index = (t_index + k_index) % len(alfabeto)
            resultado += alfabeto[nuevo_index]
        else:
            resultado += t
    return resultado

def descifrar_vigenere(texto_cifrado, clave):
    clave_ext = expandir_clave(clave, len(texto_cifrado))
    resultado = ""

    for c, k in zip(texto_cifrado, clave_ext):
        if c in alfabeto:
            c_index = alfabeto.index(c)
            k_index = alfabeto.index(k)
            nuevo_index = (c_index - k_index) % len(alfabeto)
            resultado += alfabeto[nuevo_index]
        else:
            resultado += c
    return resultado

# -----------------------------
# 3. FRASES A CIFRAR
# -----------------------------
frases = [
    "el conocimiento es poder",
    "la paciencia es la clave del exito",
    "nunca te rindas ante los desafios",
    "aprender nunca esta de mas",
    "la practica hace al maestro",
    "cada error es una oportunidad",
    "el futuro pertenece a quienes creen en sus sueños",
    "la disciplina supera al talento",
    "los grandes logros requieren tiempo",
    "confia en el proceso"
]


claves = ["love", "mpl"]


data = []

for clave in claves:
    for frase in frases:
        cifrado = cifrar_vigenere(frase, clave)
        descifrado = descifrar_vigenere(cifrado, clave)
        data.append([frase, clave, cifrado, descifrado])

df = pd.DataFrame(data, columns=[
    "Texto Original", "Clave", "Texto Cifrado", "Texto Descifrado"
])


df
```
