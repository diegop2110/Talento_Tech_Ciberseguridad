# Numeros primos encontrados de manera tradicional

```python
def numero_primo(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

primos = []

numero = 2
while len(primos) < 100:
    if numero_primo(numero):
        primos.append(numero)
    numero += 1

print(primos)

for i, primo in enumerate(primos):
    print(f" {i+1}: numero primo  {primo}")
```

#100 numeros primos con lambda

```python
numero_primo = lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) +1))

numeros = range(2, 1000)

primos = list(filter(numero_primo, numeros))[:100]

for i, primo in enumerate(primos):
    print(f"{i+1}: {primo}")

```
