```python
import pandas as pd
import random

# Alfabeto en español incluyendo la ñ
alfabeto = "abcdefghijklmnñopqrstuvwxyz"

# Cifrado César adaptado al alfabeto español
def cifrar_cesar_es(texto, desplazamiento):
    resultado = ""
    for caracter in texto:
        if caracter.lower() in alfabeto:
            es_mayus = caracter.isupper()
            indice = alfabeto.index(caracter.lower())
            nuevo_indice = (indice + desplazamiento) % len(alfabeto)
            nuevo_caracter = alfabeto[nuevo_indice]
            resultado += nuevo_caracter.upper() if es_mayus else nuevo_caracter
        else:
            resultado += caracter
    return resultado

def descifrar_cesar_es(texto_cifrado, desplazamiento):
    return cifrar_cesar_es(texto_cifrado, -desplazamiento)

# Lista de frases de ejemplo
frases = [
    "El conocimiento es poder.",
    "La paciencia es la clave del éxito.",
    "Nunca te rindas ante los desafíos.",
    "Aprender nunca está de más.",
    "La práctica hace al maestro.",
    "Cada error es una oportunidad.",
    "El futuro pertenece a quienes creen en sus sueños.",
    "La disciplina supera al talento.",
    "Los grandes logros requieren tiempo.",
    "Confía en el proceso."
]

# Generar claves aleatorias únicas
claves = random.sample(range(1, 27), len(frases))

# Construcción del DataFrame
data = []

for i in range(len(frases)):
    texto_original = frases[i]
    desplazamiento = claves[i]
    texto_cifrado = cifrar_cesar_es(texto_original, desplazamiento)
    texto_descifrado = descifrar_cesar_es(texto_cifrado, desplazamiento)
    data.append([texto_original, desplazamiento, texto_cifrado, texto_descifrado])

df = pd.DataFrame(data, columns=["Texto Original", "Clave", "Texto Cifrado", "Texto Descifrado"])

# Mostrar DataFrame final

df
```

# **¿Cuánto tiempo tardaría descifrar el mensaje usando un computador?**

Un computador puede probar todas las 27 posibles claves 0 a 26 muy rápidamente. Esto se llama un ataque por fuerza bruta. Dado que el espacio de claves es tan pequeño, incluso una computadora muy básica puede:

Probar todas las claves.

Evaluar cuál resultado tiene sentido (por ejemplo, usando diccionarios de palabras).

Un script de Python puede hacerlo en milisegundos.

# **¿Cuánto tiempo tardaría descifrar el mensaje a un grupo de personas?**
Un grupo de personas podría:

Probar manualmente cada uno de los 27 desplazamientos.

Reconocer rápidamente palabras comunes como "el", "la", "de", etc.

Si tienen papel y lápiz o una tabla alfabética, podrían encontrar la clave correcta en menos de una hora, muchas veces en menos de 10–15 minutos.

# **¿Es un método seguro para comunicar datos?**
No es un metodo seguro porque:
Es vulnerable al ataque de fuerza bruta (solo hay 27 claves posibles).

También vulnerable a análisis de frecuencia, ya que en español, letras como e, a, o aparecen con más frecuencia, lo cual ayuda a deducir el cifrado.

No hay ninguna clave secreta que sea difícil de adivinar: es solo un número.

# **¿Cómo se puede mejorar el sistema para hacerlo más seguro?**

Opciones para mejorarlo:
Usar cifrado con clave más compleja, como:

Vigenère (clave como palabra o frase que se repite)

AES / RSA (cifrado moderno de clave pública o simétrica)

Aumentar el tamaño del alfabeto (pero sigue siendo insuficiente sin una buena clave).

Agregar pasos adicionales:

Inversión de texto.

Codificación binaria o en base64.

Hashing (si no necesitas revertir el texto).

Combinar cifrados:

Cifrado César + transposición de letras = mucho más difícil de romper manualmente.
